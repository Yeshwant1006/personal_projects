```r
# Assuming your data frame is named 'mydata'

# Preprocess the data to handle missing values
mydata2 <- na.omit(mydata2[ , c("day", "Max.TemperatureC", "Mean.TemperatureC", "Min.TemperatureC")])

# Set up the plot window
plot(1, type = "n", xlim = c(1, nrow(mydata2)), ylim = range(mydata2$Min.TemperatureC, mydata2$Max.TemperatureC),
     xlab = "Day", ylab = "Temperature (°C)", main = "Temperature Variation")

# Add lines for maximum, mean, and minimum temperatures
lines(mydata2$day, mydata2$Max.TemperatureC, col = "red", lty = "dashed")
lines(mydata2$day, mydata2$Mean.TemperatureC, col = "blue")
lines(mydata2$day, mydata2$Min.TemperatureC, col = "green")

# Add a legend
legend("topright", legend = c("Max", "Mean", "Min"), col = c("red", "blue", "green"), lty = c("dashed", "solid", "solid"))

# Adjust the plot margins

par(mfrow =  c(3, 1))




```

```r

mydata2 <- mydata2[complete.cases(mydata2[, c("day", "Max.TemperatureC", "Mean.TemperatureC", "Min.TemperatureC")]), ]


plot(1, type = "n", xlim = c(1, nrow(mydata2)), ylim = range(mydata2$Min.TemperatureC, mydata2$Max.TemperatureC),
     xlab = "Day", ylab = "Temperature (°C)", main = "Temperature Variation")

lines(mydata2$day, mydata2$Max.TemperatureC, col = "yellow", lty = "dashed")
lines(mydata2$day, mydata2$Mean.TemperatureC, col = "blue")
lines(mydata2$day, mydata2$Min.TemperatureC, col = "green")


legend("topleft", legend = c("Max", "Mean", "Min"), col = c("yellow", "blue", "green"), lty = c("dashed", "solid", "solid"))


par(mfrow =  c(3, 1))
plot(mydata2[ ,2])
```
